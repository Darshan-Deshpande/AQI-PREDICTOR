# DATA EXTRACTION

```python
import requests
import json
import pandas as pd
# Adding libraries
```

## Extracting States' Codes and Names of the USA

```python
r = requests.get('https://aqs.epa.gov/data/api/list/states?email=adarshbhardwaj59@gmail.com&key=indigoosprey62')
r_json=json.loads(r.text)
# Accessing the codes and names of all states of USA from the USA EPA website
```

```python
state_df = pd.DataFrame(r_json['Data'])
# Creating a dataframe which consists of names and codes of all states of the USA
```

```python
state_df
```

## Extracting Counties' Codes and Names of all States of the USA

```python
b = []
for i in state_df['code']:
    if i != 'CC':
        p = {'state' : i}
        r = requests.get('https://aqs.epa.gov/data/api/list/countiesByState?email=adarshbhardwaj59@gmail.com&key=indigoosprey62', p)
        r_json=json.loads(r.text)
        b.append(r_json)

# Accessing the codes of all counties of all states of the USA from the USA EPA website by passing a single state at a time
```

```python
county = {}
for i in range(len(state_df)):
    if state_df.iloc[i, 0] != 'CC':
        county[state_df.iloc[i, 0]] = pd.DataFrame(b[i]['Data'])
        
# Creating a dictionary of dataframes named 'County' which consists of states' codes as keys and a dataframe,
# which consists of names and codes of counties of that particular state, as elements.
```

## Defining Necessary Functions

```python
def pollutant_data(state_code, county_list, bdate, edate, pollutant):
    b = []
    for i in county[state_code]['code']:
        p = {'param' : pollutant, 'bdate' : bdate, 'edate' : edate, 'state' : state_code, 'county' : i}
        r = requests.get('https://aqs.epa.gov/data/api/dailyData/byCounty?email=adarshbhardwaj59@gmail.com&key=indigoosprey62', p)
        r_json=json.loads(r.text)
        if r_json.get('Data') != []:
            b.append(r_json)
    return b

# The function named 'pollutant_data' is used to get AQI data for a pollutant of a state of the USA.
# In this function, 5 parameters were passed - state code, 'county' dictionary, bdate - starting date, edate - ending date, and pollutant's code.
```

```python
def final_list(data):
    final_data = []
    for i in range(len(data)):
        for j in range(len(data[i])):
            for k in range(len(data[i][j]['Data'])):
                final_data.append(data[i][j]['Data'][k])
    return final_data

# This function creates a list of dictionaries.
# From the 'pollutant_data' function, a list had obtained, but, it also consists of unnecessary data in JSON type, so, this function removes unnecessary data and returns a list of dictionaries.
```

## Extracting AQI data of CO

```python
US_CO_2019 = []
for i in range(len(state_df)):
    if state_df.iloc[i, 0] != 'CC':
        state_data = pollutant_data(state_df.iloc[i, 0], county, '20190101', '20191231', '42101')
        US_CO_2019.append(state_data)
        
# This loop extraxt the AQI data of CO for 2019, of all the states of the USA by passing one state at a time to the function named 'pollutant_data'
```

```python
US_CO_2019_final = final_list(US_CO_2019)
# Since the data obtained from the website is of JSON type, so, 'final_list' function is used to extraxt the important data in form of list of list of dictionaries.
```

```python
US_CO_2019_df = pd.DataFrame(US_CO_2019_final)
# The list of dictionaries was converted to dataframe
```

```python
US_CO_2019_df.to_csv('US_CO_2019.csv')
# Saves the raw data in form of csv file
```
